```python
def decimal_to_base(n: int, base: int) -> str:

    if not (2 <= base <= 16):
        raise ValueError("Основание системы счисления должно быть в диапазоне от 2 до 16")

    if n == 0:
        return "0"

    digits = "0123456789ABCDEF"
    result = ""
    num = abs(n)

    while num > 0:
        remainder = num % base
        result = digits[remainder] + result
        num //= base

    if n < 0:
        result = "-" + result
    return result


def base_to_decimal(s: str, base: int) -> int:

    if not (2 <= base <= 16):
        raise ValueError("Основание системы счисления должно быть в диапазоне от 2 до 16")

    digits = "0123456789ABCDEF"
    s = s.upper().strip()

    sign = 1
    if s[0] == '-':
        sign = -1
        s = s[1:]

    result = 0
    for char in s:
        if char not in digits[:base]:
            raise ValueError(f"Символ '{char}' не допустим для системы с основанием {base}")
        result = result * base + digits.index(char)

    return sign * result


if __name__ == "__main__":
    print("Перевод из 10-й системы в N-ю")
    num = int(input("Введите число в 10-й системе: "))
    base = int(input("Введите целевую систему счисления (2-16): "))
    converted = decimal_to_base(num, base)
    print(f"{num} (10) = {converted} ({base})\n")

    print("Перевод из N-й системы в 10-ю")
    num_str = input("Введите число в N-й системе: ")
    base2 = int(input("Введите основание системы (2-16): "))
    decimal_val = base_to_decimal(num_str, base2)
    print(f"{num_str} ({base2}) = {decimal_val} (10)")
```
