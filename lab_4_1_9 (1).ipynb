```python
"""
Модуль для демонстрации работы с классами, объектами и обработкой исключений.
Лабораторная работа № 4.1. Вариант 9 (Маркетинговое агентство).
"""


class Campaign:
    """
    Класс, описывающий маркетинговую кампанию (Сущность).

    Атрибуты:
        campaign_id (int): Уникальный идентификатор кампании.
        name (str): Название кампании.
        budget (float): Бюджет кампании в рублях.
    """

    def __init__(self, campaign_id: int, name: str, budget: float):
        """
        Конструктор класса Campaign с валидацией данных.
        """
        self.campaign_id = campaign_id
        self.name = name

        # РЕАЛИЗАЦИЯ: Валидация бюджета
        try:
            parsed_budget = float(budget)
            if parsed_budget < 0:
                raise ValueError("Бюджет не может быть отрицательным числом")
            self.budget = parsed_budget
        except (ValueError, TypeError) as e:
            # Если ошибка вызвана нашим raise ValueError, пробрасываем её дальше
            if "отрицательным" in str(e):
                raise ValueError(f"Некорректный бюджет для кампании '{name}': {e}")
            # Иначе это ошибка преобразования типа (например, передали строку "миллион")
            raise TypeError(f"Бюджет должен быть числом. Получен тип: {type(budget).__name__}")

    def display_info(self):
        """Метод для вывода данных атрибутов в консоль."""
        print(f"[Campaign ID {self.campaign_id}] {self.name} | Бюджет: {self.budget} руб.")

    def __str__(self):
        """Магический метод для красивого строкового представления (Vibe Code)."""
        return f"Кампания: \"{self.name}\" | Бюджет: {self.budget:,.2f} руб."


class Client:
    """
    Класс, описывающий клиента маркетингового агентства (Субъект).

    Атрибуты:
        client_id (int): Уникальный идентификатор клиента.
        company_name (str): Название компании клиента.
        brand (str): Бренд клиента.
    """

    def __init__(self, client_id: int, company_name: str, brand: str):
        """
        Конструктор класса Client с валидацией данных.
        """
        # РЕАЛИЗАЦИЯ: Валидация ID клиента
        try:
            self.client_id = int(client_id)
        except (ValueError, TypeError):
            raise TypeError(f"ID клиента должен быть целым числом. Получен тип: {type(client_id).__name__}")

        self.company_name = company_name
        self.brand = brand

    def display_info(self):
        """Метод для вывода данных атрибутов в консоль."""
        print(f"[Client ID {self.client_id}] {self.company_name} | Бренд: {self.brand}")

    def __str__(self):
        """Магический метод для красивого строкового представления (Vibe Code)."""
        return f"Клиент: {self.company_name} | Бренд: \"{self.brand}\""


if __name__ == "__main__":
    print("="*50)
    print("ДЕМОНСТРАЦИЯ КОРРЕКТНОЙ РАБОТЫ КЛАССОВ")
    print("="*50)

    # Создание корректных объектов
    campaign1 = Campaign(1, "Летний промо", 500000.0)
    campaign2 = Campaign(2, "Новогодняя распродажа", 1250000.50)

    client1 = Client(101, "ООО Техномир", "TechWorld")
    client2 = Client(102, "ИП Иванов", "Уютный дом")

    # Демонстрация работы метода __str__ (Vibe Code)
    print("\n--- Маркетинговые кампании ---")
    print(campaign1)
    print(campaign2)

    print("\n--- Клиенты агентства ---")
    print(client1)
    print(client2)

    print("\n" + "="*50)
    print("ДЕМОНСТРАЦИЯ ОБРАБОТКИ ИСКЛЮЧЕНИЙ")
    print("="*50)

    # Тест 1: Попытка передать строку вместо числа в бюджет
    print("\n1. Попытка создать кампанию с некорректным типом бюджета (строка):")
    try:
        bad_campaign = Campaign(99, "Сломанная кампания", "сто тысяч рублей")
    except TypeError as e:
        print(f"Исключение успешно перехвачено: {e}")
    except Exception as e:
        print(f"Неожиданная ошибка: {e}")

    # Тест 2: Попытка передать отрицательное число в бюджет
    print("\n2. Попытка создать кампанию с отрицательным бюджетом:")
    try:
        negative_campaign = Campaign(98, "Негативная", -5000)
    except ValueError as e:
        print(f"Исключение успешно перехвачено: {e}")
    except Exception as e:
        print(f"Неожиданная ошибка: {e}")

    # Тест 3: Попытка передать строку вместо числа в ID клиента
    print("\n3. Попытка создать клиента с некорректным ID (список вместо int):")
    try:
        bad_client = Client([1, 2], "Bad Company", "NoBrand")
    except TypeError as e:
        print(f"Исключение успешно перехвачено: {e}")
    except Exception as e:
        print(f"Неожиданная ошибка: {e}")

    print("\n Программа завершила работу корректно, не упав из-за ошибок!")
```
