```python
# Лабораторная работа № 4.1.
# Вариант 9 (Маркетинговое агентство).

class Campaign:

    def __init__(self, campaign_id: int, name: str, budget: float):

        self.campaign_id = campaign_id
        self.name = name
        self.budget = budget

    def display_info(self):

        print(f"[Campaign ID {self.campaign_id}] {self.name} | Бюджет: {self.budget} руб.")

    def __str__(self):

        return f"Кампания: \"{self.name}\" | Бюджет: {self.budget:,.2f} руб."


class Client:


    def __init__(self, client_id: int, company_name: str, brand: str):

        self.client_id = client_id
        self.company_name = company_name
        self.brand = brand

    def display_info(self):

        print(f"[Client ID {self.client_id}] {self.company_name} | Бренд: {self.brand}")

    def __str__(self):

        return f"Клиент: {self.company_name} | Бренд: \"{self.brand}\""


if __name__ == "__main__":

    campaign1 = Campaign(1, "Летний промо", 500000.0)
    campaign2 = Campaign(2, "Новогодняя распродажа", 1250000.50)
    campaign3 = Campaign(3, "Запуск нового продукта", 300000.0)

    client1 = Client(101, "ООО Техномир", "TechWorld")
    client2 = Client(102, "ИП Иванов", "Уютный дом")
    client3 = Client(103, "АО Спортмастер", "Sportmaster")

    print("--- Маркетинговые кампании ---")
    print(campaign1)
    print(campaign2)
    print(campaign3)

    print("\n--- Клиенты агентства ---")
    print(client1)
    print(client2)
    print(client3)

    print("\n--- Демонстрация метода display_info() ---")
    campaign1.display_info()
    client1.display_info()
```
