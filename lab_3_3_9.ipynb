```python
# Задание task_03_03_09.
# Выполнила: Ермакова Арина
# Группа: ЦИБ-251

"""
Ошибки (номера строк через пробел, данная строка - №2): 15 16 25
"""

def f(x):

    return x**2 / (x + 2) - 3

try:
    k = int(input("Введите границу интервала [-k; k]: "))
    h = float(input("Введите шаг табуляции: "))
except ValueError:
    print("Ошибка: введены некорректные данные. Ожидается число.")
    exit()

x = -k
print("{:>10} {:>10}".format("x", "f(x)"))
while x <= k:
    try:
        print("{:10.2f} {:10.2f}".format(x, f(x)))
    except ZeroDivisionError:
        print("{:10.2f} {:>10}".format(x, "-"))
    x += h
```
